```python
# Count the number of crashes per city and weather
crash_counts = (
    combined_crash.groupby(['City_Town_Name', 'Grouped_Weather'])
    .size()
    .reset_index(name='Crash_Count')
)
```

```python
# Group the original data
grouped = combined_crash.groupby(["City_Town_Name", "Grouped_Weather", "Crash_Severity"]).size().reset_index(name="Count")

# City dropdown filter
city_dropdown = alt.binding_select(
    options=sorted(grouped["City_Town_Name"].dropna().unique()),
    name="Select City: "
)
city_select = alt.selection_point(
    fields=["City_Town_Name"],
    bind=city_dropdown,
    value="SOMERVILLE"  
)


```

```python
# Heatmap with Grouped_Weather vs Crash Severity
heatmap = alt.Chart(grouped).mark_rect().encode(
    x=alt.X('Crash_Severity:N', title='Crash Severity'),
    y=alt.Y('Grouped_Weather:N', title='Weather Condition'),
    color=alt.Color('Count:Q', title='Crash Count', scale=alt.Scale(scheme='tealblues')),
    tooltip=['City_Town_Name', 'Grouped_Weather', 'Crash_Severity', 'Count']
).add_params(
    city_select
).transform_filter(
    city_select
).properties(
    title='Heatmap: Weather vs. Crash Severity (Filter by City)',
    width=600,
    height=500
)

heatmap
```

```python
# Groups and counts crashes by city and intersection
intersection_counts = (
    combined_crash.groupby(['City_Town_Name', 'At_Roadway_Intersection'])
    .size()
    .reset_index(name='Crash_Count')
)

# Keep only the top 5 intersections per city
top_5_each_city = (
    intersection_counts
    .sort_values(['City_Town_Name', 'Crash_Count'], ascending=[True, False])
    .groupby('City_Town_Name')
    .head(5)
)

# Generate a chart for each city
city_charts = []

for city in top_5_each_city['City_Town_Name'].unique():
    chart = (
        alt.Chart(top_5_each_city[top_5_each_city['City_Town_Name'] == city])
        .mark_bar()
        .encode(
            y=alt.Y('At_Roadway_Intersection:N', sort='-x', title='Intersection'),
            x=alt.X('Crash_Count:Q', title='Number of Crashes'),
            tooltip=['At_Roadway_Intersection', 'Crash_Count']
        )
        .properties(
            title=f'Top 5 Intersections in {city}',
            width=500,
            height=200
        )
    )
    city_charts.append(chart)

# Display all charts vertically
alt.vconcat(*city_charts)
```

```python
# Parse crash time and extract hour
combined_crash['Crash_Time_Parsed'] = pd.to_datetime(combined_crash['Crash_Time'], format='%I:%M %p', errors='coerce')
combined_crash['Crash_Hour'] = combined_crash['Crash_Time_Parsed'].dt.hour
```

```python
# Group by hour, crash type, and city
crash_hourly = (
    combined_crash
    .groupby(['Crash_Hour', 'Manner_of_Collision', 'City_Town_Name'])
    .size()
    .reset_index(name='Crash_Count')
)
```

```python
# City dropdown selection
city_dropdown = alt.binding_select(
    options=sorted(crash_hourly['City_Town_Name'].dropna().unique()),
    name='Select City: '
)
city_select = alt.selection_point(
    fields=['City_Town_Name'],
    bind=city_dropdown,
    value=sorted(crash_hourly['City_Town_Name'].dropna().unique())[0]  # default city
)

# Crash type dropdown selection
type_dropdown = alt.binding_select(
    options=sorted(crash_hourly['Manner_of_Collision'].dropna().unique()),
    name='Crash Type: '
)
type_select = alt.selection_point(
    fields=['Manner_of_Collision'],
    bind=type_dropdown,
    value='Rear-end' 
)
```

```python
# Hour slider
hour_slider = alt.binding_range(min=0, max=23, step=1, name='Min Hour: ')
hour_param = alt.param(name='HourFilter', bind=hour_slider, value=7)
```

```python
# Chart to show crash frequency by crash type per hour 
line = alt.Chart(crash_hourly).mark_line(point=True).encode(
    
    # X-axis: Hour of the day (formatted into AM/PM for readability)
    x=alt.X('Crash_Hour:O', title='Hour of Day',
        axis=alt.Axis(
            labelExpr="""datum.label == '0' ? '12:00 AM' :
                         datum.label < 12 ? datum.label + ':00 AM' :
                         datum.label == '12' ? '12:00 PM' :
                         (datum.label - 12) + ':00 PM'"""
        )
    ),

    # Y-axis: Total crash count
    y=alt.Y('Crash_Count:Q', title='Crash Count'),

    # Line color distinguishes crash type (e.g., angle, rear-end, sideswipe)
    color=alt.Color('Manner_of_Collision:N', title='Crash Type'),

    # Tooltip shows detailed info on hover
    tooltip=[
        alt.Tooltip('Crash_Hour:O', title='Hour'),
        alt.Tooltip('Manner_of_Collision:N', title='Crash Type'),
        alt.Tooltip('City_Town_Name:N', title='City'),
        alt.Tooltip('Crash_Count:Q', title='Crash Count')
    ]
)

# Add interactivity: filter by hour, city, and crash type
line = line.add_params(
    hour_param,     
    city_select,   
    type_select    
).transform_filter(
    city_select
).transform_filter(
    type_select
).transform_filter(
    'datum.Crash_Hour >= HourFilter'  
).properties(
    title='Crash Types by Hour of Day',
    width=700,
    height=400
)

line

```
